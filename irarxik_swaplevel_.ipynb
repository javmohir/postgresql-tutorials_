```python
df=pd.read_csv("/content/Practicum_Datasets/world_cities_air_water_quality.csv")
df
```

```python
columns=['City','Region','Country','AirQuality','WaterPollution']
df.columns=columns
df
```

```python
df['Region']=df['Region'].map(lambda x: x.replace('"',""))  # Region ustuniga tegishli ma'lum otlardan " belgisini olib tashlash
df
```

```python
df['Country']=df['Country'].map(lambda x:x.replace('"',"") ) # Country ustuniga tegishli ma'lumotlardan " belgisini olib tashlash
df
```

```python
df.to_csv("Updated_dataset.csv", index=False)
```
