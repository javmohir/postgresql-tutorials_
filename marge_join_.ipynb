```python
A=pd.DataFrame({'transport':['poezd', 'avtomabil', 'kema'],
                'tezligi':['70', '180', '45']})
A
```

```python
B=pd.DataFrame({'transport':['poezd', 'avtomabil', 'tramvay'],
                'narxi':['98000', '10000', '65000']})
B
```

```python
marged=pd.merge(A,B, how='inner')  # inner join - ichki
marged
```

```python
mar_out=pd.merge(A,B, how='outer') # outer join - tashqi
mar_out
```

```python
m_left=pd.merge(A,B, how='left')  # left join - chap
m_left
```

```python
m_right=pd.merge(A,B, how='right')  # right join - chap
m_right
```

```python
C = pd.DataFrame({'Transport_turi': ['poezd', 'avtomabil', 'kema'],
                  'Tezligi':['70', '180', '45']})

C
```
