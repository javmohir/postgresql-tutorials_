Ba'zida, tadqiqod uchun jadvaldagi qiymatlar o'rnini aralashtirib tashlash talab qilinishi mumkin. Misol uchun, sizda ma'lumotlar ma'lum bir tartibda turibdi, lekin siz ma'lumotlarni tasodifiy tartibda chiqarmoqchiisiz, buning uchun numpy kutubxonasining random modulidagi permutation funksiyasiga murojaat qilamiz.

```python
df
```

Jadvalimizda 93 ta qator mavjud, biz 1-93 gacha sonlarni tasodifiy tartibda saqlovchi ro'yxat yaratamiz:

```python
rlist=np.random.permutation(93)
rlist
```

Navbat, df ni yuqoridagi tasodifiy tartibda tahlash, buning uchun .take() meotidga murojaat qilamiz



```python
df.take(rlist)
```
